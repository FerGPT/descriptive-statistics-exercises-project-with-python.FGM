````python
# Chi**2
import numpy as np
import matplotlib.pyplot as plt

# Degrees of freedom
df = 3

# Generate a Chi-square distribution with 100 elements
chi = np.random.chisquare(df, 100)

# Plotting the Chi-square distribution
plt.figure(figsize=(10, 6))
plt.hist(chi, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Chi-square Distribution with 3 Degrees of Freedom')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````
